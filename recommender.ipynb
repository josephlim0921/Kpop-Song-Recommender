# Data Preparation

```python
indices = training_df[['track', 'artist']].reset_index()
print(indices)
```

```python
training_df = training_df[['track','danceability','energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness','instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']]
training_df
```

```python
training_df = training_df.set_index('track')
training_df
```

# Recommendation Engine

```python
# Choose a song that I want to find similarities for

desired_song = indices[(indices['artist'] == 'NewJeans') & (indices['track'] == 'Cool With You')]
desired_song
```

```python
from sklearn.preprocessing import StandardScaler

# I need to standardize the dataset

training_df_std = StandardScaler().fit_transform(training_df.values)
training_df_std.shape
```

```python
sample = training_df_std[desired_song['index']]
sample = sample.reshape(1,12)
sample
```

```python
from sklearn.metrics.pairwise import cosine_similarity

scores = cosine_similarity(training_df_std, sample)
score_df = pd.DataFrame(scores, columns=['score'])
score_df = score_df.sort_values(by='score', ascending=False).reset_index()
combined = score_df.merge(indices, on='index')
combined = combined[['track', 'artist','score','index']]
combined.head(20)
```
